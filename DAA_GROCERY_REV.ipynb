```python
# Fungsi untuk menghitung penggunaan memori
def memory_usage():
    process = psutil.Process(os.getpid())
    return process.memory_info().rss / (1024 ** 2)  # Mengembalikan dalam MB

# Membaca data dari file yang sudah dibersihkan
file_path = 'Cleaned_GroceryDataset.csv'  # File cleaned data
df = pd.read_csv(file_path)

# Membersihkan kolom 'Sub Category'
df.loc[:, 'Sub Category'] = df['Sub Category'].str.strip()  # Hapus spasi kosong di awal dan akhir

# Menampilkan kategori yang tersedia
print("Kategori yang tersedia:", df['Sub Category'].unique())

# Memulai timer untuk menghitung durasi dari input ke output
user_budget = int(input("Masukkan budget Anda: "))
user_category_input = input("Masukkan jenis makanan yang ingin Anda cari (contoh: Bakery & Desserts, Snacks): ").strip()

# Mengukur penggunaan memori sebelum proses
memory_before = memory_usage()

start_time = time.time()  # Timer dimulai setelah input kategori

# Memproses input kategori
user_categories = [category.strip() for category in user_category_input.split(',')]

# Memfilter data berdasarkan kategori pengguna
filtered_df = df[df['Sub Category'].str.contains('|'.join(user_categories), case=False, na=False)]

# Fungsi greedy untuk memilih item dalam anggaran
def knapsack_greedy(filtered_df, budget):
    """
    Fungsi untuk memilih item menggunakan metode greedy.

    Parameters:
        filtered_df (DataFrame): Data yang sudah difilter.
        budget (float): Budget maksimum.

    Returns:
        List item yang dipilih sesuai anggaran.
    """
    sorted_df = filtered_df.sort_values(by='Price', ascending=True)
    selected_items = []
    total_price = 0.0

    for _, row in sorted_df.iterrows():
        if total_price + row['Price'] <= budget:
            selected_items.append((row['Title'], row['Price']))
            total_price += row['Price']
        else:
            break

    return selected_items

# Menampilkan hasil berdasarkan kategori dan budget
if filtered_df.empty:
    print("Tidak ada item yang sesuai dengan kategori tersebut.")
else:
    result = knapsack_greedy(filtered_df, user_budget)

    if result:
        total_price = sum(price for title, price in result)
        print("\nItem yang bisa dibeli:")
        for title, price in result:
            print(f"- {title}: ${price}")
        print(f"\nTotal harga: ${total_price:.2f}")
    else:
        print("Anggaran Anda tidak mencukupi untuk membeli item dalam kategori ini.")

# Menghentikan timer dan menghitung durasi
end_time = time.time()
execution_time = end_time - start_time

# Mengukur penggunaan memori setelah proses
memory_after = memory_usage()

# Menampilkan durasi eksekusi dan penggunaan memori
print(f"\nWaktu eksekusi: {execution_time:.2f} detik")
print(f"Penggunaan memori sebelum eksekusi: {memory_before:.2f} MB")
print(f"Penggunaan memori setelah eksekusi: {memory_after:.2f} MB")
print(f"Penggunaan memori selama eksekusi: {memory_after - memory_before:.2f} MB")
```

```python
# Fungsi untuk menghitung penggunaan memori
def memory_usage():
    process = psutil.Process(os.getpid())
    return process.memory_info().rss / (1024 ** 2)  # Mengembalikan dalam MB

# Fungsi Knapsack menggunakan Dynamic Programming
def knapsack(df, budget):
    # Ambil data harga dan indeks
    prices = df['Price'].astype(int).tolist()  # Konversi harga menjadi bilangan bulat
    titles = df['Title'].tolist()
    n = len(prices)

    # Inisialisasi DP array
    dp = [[0] * (budget + 1) for _ in range(n + 1)]

    # Knapsack algorithm
    for i in range(1, n + 1):
        for w in range(budget + 1):
            if prices[i - 1] <= w:
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - prices[i - 1]] + prices[i - 1])
            else:
                dp[i][w] = dp[i - 1][w]

    # Lacak item yang dipilih
    w = budget  # Inisiasi 'w' dengan nilai budget
    selected_items = []  # List kosong untuk menyimpan barang yang dipilih
    for i in range(n, 0, -1):  # Lacak item dari akhir ke awal
        if dp[i][w] != dp[i - 1][w]:
            selected_items.append((titles[i - 1], prices[i - 1]))  # Tambahkan barang ke daftar
            w -= prices[i - 1]  # Kurangi kapasitas anggaran dengan harga barang yang dipilih

    return selected_items

# Membaca file cleaned dataset
file_path = 'Cleaned_GroceryDataset.csv'  # File cleaned data
df = pd.read_csv(file_path)

# Membersihkan kolom 'Sub Category'
df.loc[:, 'Sub Category'] = df['Sub Category'].str.strip()  # Hapus spasi kosong di awal dan akhir

# Menampilkan kategori yang tersedia
print("Kategori yang tersedia:", df['Sub Category'].unique())

# Input dari pengguna
user_budget = int(input("Masukkan budget Anda: "))
user_category_input = input("Masukkan jenis makanan yang ingin Anda cari (contoh: Bakery & Desserts, Snacks): ").strip()

# Mengukur penggunaan memori sebelum proses
memory_before = memory_usage()

# Timer dimulai setelah input kategori
start_time = time.time()

# Split kategori yang dimasukkan oleh pengguna dan hilangkan spasi berlebih
user_categories = [category.strip() for category in user_category_input.split(',')]

# Filter data berdasarkan kategori yang dimasukkan pengguna
filtered_df = df[df['Sub Category'].str.contains('|'.join(user_categories), case=False, na=False)]

if filtered_df.empty:
    print("Tidak ada item yang sesuai dengan kategori tersebut.")
else:
    # Panggil fungsi knapsack
    result = knapsack(filtered_df, user_budget)

    # Timer dihentikan setelah hasil selesai
    end_time = time.time()
    execution_time = end_time - start_time

    # Mengukur penggunaan memori setelah proses
    memory_after = memory_usage()

    # Tampilkan hasil
    if result:
        total_price = sum(price for title, price in result)  # Menghitung total harga
        print("\nItem yang bisa dibeli:")
        for title, price in result:
            print(f"- {title}: ${price}")
        print(f"\nTotal harga: ${total_price}")
    else:
        print("Anggaran Anda tidak mencukupi untuk membeli item dalam kategori ini.")

    # Tampilkan waktu eksekusi
    print(f"\nWaktu eksekusi: {execution_time:.2f} detik")

    # Tampilkan penggunaan memori
    print(f"Penggunaan memori sebelum eksekusi: {memory_before:.2f} MB")
    print(f"Penggunaan memori setelah eksekusi: {memory_after:.2f} MB")
    print(f"Penggunaan memori selama eksekusi: {memory_after - memory_before:.2f} MB")
```

```python
# Fungsi untuk menghitung penggunaan memori
def memory_usage():
    process = psutil.Process(os.getpid())
    return process.memory_info().rss / (1024 ** 2)  # Mengembalikan dalam MB

# Fungsi Backtracking
def find_parcel(items, prices, budget, current_parcel=[], current_sum=0, index=0):
    if current_sum > budget:  # Jika total harga melebihi budget, hentikan pencarian
        return []

    if index == len(items):  # Jika sudah memeriksa semua barang, kembalikan hasil
        return [current_parcel]

    # Backtracking: Dengan atau tanpa item saat ini
    with_current = find_parcel(
        items, prices, budget,
        current_parcel + [(items[index], prices[index])],
        current_sum + prices[index],
        index + 1
    )

    without_current = find_parcel(
        items, prices, budget,
        current_parcel,
        current_sum,
        index + 1
    )

    # Pilih hasil dengan total harga tertinggi
    sum_with_current = sum(price for _, price in with_current[0]) if with_current else 0
    sum_without_current = sum(price for _, price in without_current[0]) if without_current else 0

    if sum_with_current > sum_without_current:
        return with_current
    return without_current

# Membaca dataset
file_path = 'Cleaned_GroceryDataset.csv'  # Pastikan path file Anda benar
df = pd.read_csv(file_path)

# Membersihkan spasi pada kolom Sub Category
df.loc[:, 'Sub Category'] = df['Sub Category'].str.strip()

# Menampilkan kategori yang tersedia
print("Kategori yang tersedia:", df['Sub Category'].unique())

# Input pengguna
user_budget = int(input("Masukkan budget Anda: "))
user_category_input = input("Masukkan jenis makanan yang ingin Anda cari (contoh: Bakery & Desserts, Snacks): ").strip()

# Filter berdasarkan kategori pengguna
user_categories = [category.strip() for category in user_category_input.split(',')]
filtered_df = df[df['Sub Category'].str.contains('|'.join(user_categories), case=False, na=False)]

if filtered_df.empty:
    print("Tidak ada item yang sesuai dengan kategori tersebut.")
else:
    # Urutkan data berdasarkan harga (opsional untuk efisiensi)
    filtered_df = filtered_df.sort_values(by='Price')

    # Ambil data judul dan harga
    items = filtered_df['Title'].tolist()
    prices = filtered_df['Price'].astype(int).tolist()

    # Mengukur penggunaan memori sebelum proses
    memory_before = memory_usage()

    # Timer mulai
    start_time = time.time()

    # Panggil fungsi backtracking
    best_parcel = find_parcel(items, prices, user_budget)

    # Timer selesai
    end_time = time.time()
    execution_time = end_time - start_time

    # Mengukur penggunaan memori setelah proses
    memory_after = memory_usage()

    # Tampilkan hasil
    if best_parcel:
        total_price = sum(price for _, price in best_parcel[0])  # Menjumlahkan harga dari daftar item yang dipilih
        print("\nItem yang bisa dibeli:")
        for title, price in best_parcel[0]:
            print(f"- {title}: ${price}")
        print(f"\nTotal harga: ${total_price}")
    else:
        print("Anggaran Anda tidak mencukupi untuk membeli item dalam kategori ini.")

    # Waktu eksekusi
    print(f"\nWaktu eksekusi: {execution_time:.2f} detik")

    # Penggunaan memori
    print(f"Penggunaan memori sebelum eksekusi: {memory_before:.2f} MB")
    print(f"Penggunaan memori setelah eksekusi: {memory_after:.2f} MB")
    print(f"Penggunaan memori selama eksekusi: {memory_after - memory_before:.2f} MB")
```
